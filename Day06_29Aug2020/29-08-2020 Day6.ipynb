````python
# 1. read the data 
import pandas as pd
from sklearn.datasets import load_iris
iris_data = load_iris()
iris_data
````

````python
# 2.Check the null values
#3.Seperate the input and output labels
x = iris_data.data
# x = iris_data['data']
x
````

````python
y = iris_data['target']
y
````

````python
# 4.Seperate the data into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,
                                                 test_size=0.25,
                                                 random_state=1)
````

````python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
````

````python
# 5. train the model by using fit method
tree.fit(x_train,y_train)
````

````python
# it is used for downloading the tree into our local system
from sklearn.tree import export_graphviz
export_graphviz(tree,out_file='iris_tree.dot',
                filled = True,
               rounded = True,
               feature_names = iris_data.feature_names,
               class_names = iris_data.target_names)
````

````python
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
sio = StringIO()
export_graphviz(tree,out_file = sio,
                filled = True,
                rounded= True,
               feature_names=iris_data.feature_names,
               class_names=iris_data.target_names)
pplot = pydotplus.graph_from_dot_data(sio.getvalue())
````
