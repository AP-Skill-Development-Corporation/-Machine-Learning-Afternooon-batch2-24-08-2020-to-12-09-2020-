## DecisionTree Regressor

````python
# 1. Read the Data 
import pandas as pd
companies_data = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/1000_Companies.csv")
companies_data.head()
````

````python
# 2. Check the any null or process the data
companies_data.isnull().sum()
````

````python
companies_data.dtypes
````

````python
companies_data['State'].unique()
````

````python
companies_data.shape
````

````python
state1 = pd.get_dummies(companies_data['State'])
state1
````

````python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
state2 = lab.fit_transform(companies_data['State'])
print(state2)
````

````python
companies_data['State_tran']=state2
companies_data
````

````python
# seperate the input and output labels
x = companies_data[['R&D Spend','Administration','Marketing Spend','State_tran']]
x
````

````python
y = companies_data['Profit']
y
````

````python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.3,random_state=1)
````

````python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
````

````python
#train the model
tree.fit(x_tr,y_tr)
````

````python
pred = tree.predict(x_te)
````

````python
tree.score(x_tr,y_tr)
````

````python
from sklearn.metrics import r2_score
r2_score(y_te,pred)
````

# Random Forest

It is ensemble model(Combination some models).It is one of the very full algorithm.It is also used for both Classification and Regression.

````python
import pandas as pd
diabets = pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Classification/diabetes.csv")
diabets.head()
````

````python
# any null values
diabets.isna().sum()
````

````python
x = diabets.drop('Outcome',axis=1)
x.head()
````

````python
y = diabets['Outcome']
y.head()
````

````python
diabets.shape
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.25)
````

````python
from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier()
````

````python
rand.fit(x_train,y_train)
````

````python
y_pred = rand.predict(x_test)
````

````python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
````

````python
confusion_matrix(y_test,y_pred)
````
